```python
df=pd.read_csv("/kaggle/input/cars-csv/Cars93.csv")
```

```python
df.rename(columns={"Luggage.room":"Luggageroom"},inplace="True")
df
```

```python
#fill null values of luggageroom column with mean
x=df["Luggageroom"].mean()
df["Luggageroom"].fillna(x,inplace=True)
df.isnull().sum()
```

```python
#fill null values of Airbags column with mode
x=df["AirBags"].mode()[0]
x
df["AirBags"].fillna(x,inplace=True)
df.isnull().sum()
```

```python
#count the duplicate rows in dataset

df.duplicated().sum()
```

```python
#drop the duplicates
df.drop_duplicates(inplace=True)
df
```
